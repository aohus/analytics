### 1. 빵집 리스트 만들기

```python
def make_bakery_list(bakery_df):
    # 저장된 빵집이름 리스트화하기(빵집리스트 생성)
    # 1) df에서 빵집이름 컬럼 선택 
    bakery_title = bakery_df.loc[:,['빵집이름']]
    print("빵집이름 전체 갯수: "+str(len(bakery_title)))

    # 2) NaN 데이터 제거
    bakery_title = bakery_title[bakery_title['빵집이름'].notnull()]
    print("null 제거: "+str(len(bakery_title)))

    # 3) dataframe -> list 형변환
    bakery_list = bakery_title.빵집이름.tolist()

    # 4) 한문자열에 여러 빵집쓰여있는 것 분리 후 저장, ("어글리 건담 오월의" -> "어글리", "건담", "오월의")
    dellist = []
    for i in range(0,len(bk_list)):
        split_str = bakery_list[i].split()
        split_len = len(split_str)
        if split_len != 1:
            dellist.append(i)
            for j in range(0,split_len):
                bakery_list.append(split_str[j])
        
    for i in range(0,len(dellist)):
        del bakery_list[dellist[i]]
        
    print("빵집 리스트 분리, 추가: "+str(len(bakery_list)))
    
    # 5) 중복제거
    bakery_list = list(set(bakery_list))
    print("중복 제거: "+str(len(bakery_list)))
    
    return bakery_list
```

```python
bk6ago = pd.read_csv('../data/posts_2015.csv')
bk_list = make_bakery_list(bk6ago)
```

```python
# 명사 추출하기 위한 라이브러리 konlpy 설치
# 추출 명사 count를 위한 collection
from konlpy.tag import Okt
from collections import Counter
```

```python
# 파일 불러오기
bk6ago = pd.read_csv('../data/posts_tmp.csv')
```

```python
content = bk6ago.loc[:,"내용"]
content.head(3)
#print(okt.morphs(total[1][3]))
#print(okt.(total[1][3]))
```

```python
# okt 객체 생성
okt = Okt()
word = []

# okt 단어 사전에 추가해주기
# 명사 추출
noun = []
for i in range(0,len(content)):
    noun += okt.nouns(content[i])
```

```python
##################################
### 그냥 실험 morphs, pos 이용  #####
##################################

# 품사 포함 단어 추출
for i in range(0, len(content)):
    word += okt.morphs(content[i])
    
# 품사별 추출
pos = []
for i in range(0,len(content)):
    pos += okt.pos(content[i])

#print(word[0])
#print(noun[0])
#type(pos) : list
for i in range(20):
    if pos[i][1] != 'Josa':
        print(pos[i])
```

```python
# 명사빈도 카운트
count = Counter(noun)
noun_list = count.most_common(150)

print(noun_list)
```
